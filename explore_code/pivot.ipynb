```python
import pandas as pd
```

```python
df=pd.read_csv("../active.csv",sep=";")
```

```python
df.columns
```

```python
idpaciente=df.idpaciente.unique()
idpaciente[0]
```

```python

tipos= df.tipo.unique()
tipos
```

```python
chunck=df.query('idpaciente == @idpaciente[0] & tipo == @tipos[0]')
chunck=pd.DataFrame(chunck.loc[:,["idsintoma",'valor','data','sintoma']])
chunck.set_index(['idsintoma'], inplace=True)
chunck
```

```python
sintomas=pd.DataFrame(chunck.loc[:,["sintoma"]])
sintomas.drop_duplicates('sintoma', inplace=True)
sintomas
```

```python
datas=chunck.data.unique()
datas.sort()

```

```python

import datetime
def to_date(s):
    return datetime.datetime.strptime(s, '%Y-%m-%d').date()
```

```python
dataseries=([to_date(s) for s in chunck.data.values])
```

```python
chunck['data'] = dataseries
 
#chunck.data=pd.Series([to_date(s) for s in chunck.data.values])
chunck
```

```python
cutoff=datetime.datetime.today().date()-datetime.timedelta(days=120)
cutoff
```

```python
chunck=pd.DataFrame(chunck.query("data >= @cutoff"))
chunck
```

```python
datas=chunck.data.unique()
datas
```

```python

chunck
```

```python
out=chunck.pivot(columns='data',values='valor')
```

```python

out
```

```python
final=sintomas.merge(out,left_index=True,right_index=True)
final
```

```python
import xlsxwriter
writer = pd.ExcelWriter("pandas_datetime.xlsx",
                        engine='xlsxwriter',
                        datetime_format='mmm d yyyy hh:mm:ss',
                        date_format='mmmm dd yyyy')
```

```python
final.to_excel(writer, sheet_name='Sheet1', startrow=0, header=True,index=True)
```
